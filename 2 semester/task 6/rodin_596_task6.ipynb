**Предобработка данных**

- Убрать пользователей, которые проставили мало рейтингов из обучающей выборки.

- Убрать товары без рейтингов из выборки.

**Коллаборативная фильтрация**

Идея: разложим матрицу R на U и P, $R \approx UP$. U - эмбеддинг пользователей, P - эмбеддинг товаров. Скалярное произведения представления пользователя на представление товара дает рейтинг. Рекомендательная система выдает товары с наибольшим полученным рейтингом.

- ALS(Alternating Least Squares) - По очереди замораживаем матрицы U,P и решаем задачу оптимизации.

- SVG - раскладываем R через SVD, потом оставляем часть главных значений.

- Neural Networks - обучаем два эмбеддинга. Их выходы скалярно умножаем и считаем функцию потерь с реальным рейтингом.

- Ансамбль моделей

**Ранжирующие алгоритмы**


- Eigen Rank -  Для каждого пользователя u находим некоторую окрестность через тау корреляцию Кенделла и считаем функцию предпочтения. $\phi(i,j) = \frac{\sum_{v \in N_{u}^{i,j}} s_{u,v}(r_{u,j} - r_{u,i})}{\sum_{v \in N_{u}^{i,j}} s_{u,v}}$Если пользователи из $N_u$ присваивают i рейтинг больще j, то $\phi(i,j) > 0$ и $\phi(i,j) < 0$.

- LambdaMART - оптимизируем функцию потерь напрямую градиентным бустингом на деревьях. Так как она часто не дифференцируема, то оцениваем градиент приблизительно.


**Метрики**

- NSCG - Normalized Discounted Cumulative Gain. Суммируем прирост от релевантности, домноженный на штраф от позиции в рейтинге.

- MAP- Mean Average Precision - считаем процент релевантного от всего рекомендованного по всем пользователям.
